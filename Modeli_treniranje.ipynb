```python
import numpy as np
import tensorflow as tf
import os
from MetodeModeli import *
```

```python
BR_KLASA = 6
EPOCHS = 100
BATCH_SIZE = 16
```

```python
DIR_DATA = 'D:\\Ivana\\01 Biologijom nadahnuto računarstvo\\Podaci_train_test'
MODEL_DIR = 'MODELI\\'
TRAIN_DIR = DIR_DATA + '\\Train'
TEST_DIR = DIR_DATA + '\\Test'
```

```python
EPOCHS_FE = 25
EPOCHS_FT = 75
```

### InceptionV3 
Combined

```python
IN_SHAPE = (299, 299, 3)
```

```python
odabrani_model = tf.keras.applications.InceptionV3
preprocess_func = tf.keras.applications.inception_v3.preprocess_input 
```

```python
data_gen_train, data_gen_val = data_generators(preprocess_fun = preprocess_func, DA=True)
```

```python
tf.keras.backend.clear_session()
```

```python
train_flow = data_gen_train.flow_from_directory(TRAIN_DIR, 
                                                target_size=(IN_SHAPE[0], IN_SHAPE[1]),
                                                batch_size = BATCH_SIZE, 
                                                class_mode = 'categorical')

validation_flow = data_gen_val.flow_from_directory(TEST_DIR, 
                                                   target_size=(IN_SHAPE[0], IN_SHAPE[1]),
                                                   batch_size=BATCH_SIZE, 
                                                   class_mode='categorical', 
                                                   shuffle=False)
```

```python
train_combinedModel(odabrani_model, IN_SHAPE, 
                    train_flow, validation_flow,
                    model_dir=MODEL_DIR, 
                    label='InceptionV3_FE_FT_BN_inference2',
                    epochs_FE=EPOCHS_FE, epochs_FT=EPOCHS_FT, 
                    lr_FE =5e-4, lr_FT=5e-5)
```

```python
train_combinedModel(odabrani_model, IN_SHAPE, 
                    train_flow, validation_flow,
                    model_dir=MODEL_DIR, 
                    label='InceptionV3_FE_FT_BN_inference',
                    epochs_FE=EPOCHS_FE, epochs_FT=EPOCHS_FT, 
                    lr_FE =5e-4, lr_FT=5e-5)
```
